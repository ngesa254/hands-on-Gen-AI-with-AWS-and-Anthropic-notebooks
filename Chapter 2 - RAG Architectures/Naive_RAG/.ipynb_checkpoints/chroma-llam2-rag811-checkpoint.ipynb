## Load knowledge base

```python
loader = SimpleDirectoryReader(input_files=["../../Data/t.pdf"])

iot_docs = loader.load_data()
```

### Define LLM and Content Handler

```python
class ContentHandlerForTextGeneration(LLMContentHandler):
    content_type = "application/json"
    accepts = "application/json"

    def transform_input(self, prompt: str, model_kwargs: dict) -> bytes:
        input_str = json.dumps({"inputs": [[{"role": "user", "content": prompt},]],
                                  "parameters" : model_kwargs
                                  })
        return input_str.encode('utf-8')

    def transform_output(self, output: bytes) -> str:
        response_json = json.loads(output.read().decode("utf-8"))
        return response_json[0]['generation']['content']
```

```python
parameters = {
    #"generate_kwargs": {},
    "max_new_tokens": 1024,
    #"context_window":3900,
    "temperature": 0.1,}

region="eu-west-1"

endpoint_name = "iot-sagemaker-llama-2-llama-model-Endpoint" 
content_handler = ContentHandlerForTextGeneration()

llm = SagemakerEndpoint(
    endpoint_name=endpoint_name,
    region_name=region,
    model_kwargs=parameters,
    endpoint_kwargs={"CustomAttributes":"accept_eula=true"},
    content_handler=content_handler)
```

### Define Embedding model

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

### Vector StoreIndex

```python
from llama_index.core import Settings

Settings.llm = llm
Settings.embed_model = embed_model
```

```python
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_documents(iot_docs)
```

```python
# set Logging to DEBUG for more detailed outputs
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("Which packages do we have in Telematcis?")
display(Markdown(f"{response}"))
```

```python
query_engine = index.as_query_engine(similarity_top_k=3)
# response = query_engine.query("What was mpesa deposits in 2022")
response = query_engine.query("Write a blog to share insights about Safaricom Telematics Benefits?")
display(Markdown(f"<b>{response}</b>"))
```

```python
import chromadb
from llama_index.core import VectorStoreIndex
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core import StorageContext
```

```python
# create client
db = chromadb.PersistentClient(path="./chroma_db")
chroma_collection = db.get_or_create_collection("iot_collection")
```

```python
# save embedding to disk
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
```

```python
# create index
index = VectorStoreIndex.from_documents(
    iot_docs, storage_context=storage_context, embed_model=embed_model
)
```

#### Load from disk

```python
# load from disk
db2 = chromadb.PersistentClient(path="./chroma_db")
chroma_collection = db2.get_or_create_collection("iot_collection")
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
index = VectorStoreIndex.from_vector_store(
    vector_store,
    embed_model=embed_model,
)
```

#### Query data

```python
# Query Data from the persisted index
query_engine = index.as_query_engine(llm=llm, similarity_top_k=3)
```

```python
response = query_engine.query("Write a blog to share insights about Safaricom Telematics Benefits?")
display(Markdown(f"<b>{response}</b>"))
```
